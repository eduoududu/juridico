## Groq

```python
# Recuperar a chave da API do OpenRouter de forma segura
openrouter_api_key = userdata.get('groq_api_key')

# Inicializar o cliente OpenRouter
# OpenRouter usa a mesma API do OpenAI, mas com um base_url diferente
client = OpenAI(
    base_url="https://api.groq.com/openai/v1",
    api_key=openrouter_api_key,
)

# Listar os modelos disponíveis
models = client.models.list()

print("Modelos gratuitos disponíveis no OpenRouter:")
free_models = [model.id for model in models.data if ":free" in model.id]
for model_id in free_models:
    print(f"- {model_id}")
```
